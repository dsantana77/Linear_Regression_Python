```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# Dummies

```python
# Substitua 'caminho/do/arquivo.xlsx' pelo caminho do seu arquivo Excel
caminho_do_arquivo = 'Limite_Credito_Escolaridade.xlsx'

# Carregar o arquivo Excel em um DataFrame
limite = pd.read_excel(caminho_do_arquivo)
limite.head()
```

```python
limite = pd.get_dummies(limite, prefix=["Escolaridade"])
limite.head()
```

```python
# Converter os valores booleanos em inteiros (0 e 1)
limite = limite.astype(int)

# Agora 'limite' contém os valores 0 e 1 em vez de True e False

# Você pode imprimir o DataFrame resultante para verificar
limite.head()
```

```python
limite.corr()
```

### Ajuste de variáveis

```python
import statsmodels.formula.api as smf
```

```python
function = "LimitedoChequeEspecial ~ Idade + RendimentoTotal + Salario + LimitedeCreditoImediato + Escolaridade_Fundamental_Medio + Escolaridade_Superior_Pos"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary()) # tirar IDADE
```

```python
function = "LimitedoChequeEspecial ~ RendimentoTotal + Salario + LimitedeCreditoImediato + Escolaridade_Fundamental_Medio + Escolaridade_Superior_Pos"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary()) # tirar Intercepto acrescentar -1
```

```python
function = "LimitedoChequeEspecial ~ RendimentoTotal + Salario + LimitedeCreditoImediato + Escolaridade_Fundamental_Medio + Escolaridade_Superior_Pos - 1"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary()) # tirar Escolaridade_Superior_Pos  
```

```python
function = "LimitedoChequeEspecial ~ RendimentoTotal + Salario + LimitedeCreditoImediato + Escolaridade_Fundamental_Medio - 1"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary()) # Selecionado
```
